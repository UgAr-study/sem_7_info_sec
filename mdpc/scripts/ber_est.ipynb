```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
res_files = glob.glob('results/out-*')
df = pd.DataFrame(data = {'SNR':[], 'Errors':[], 'Total': []})
for filename in res_files:
    df_one_snr = pd.read_csv(filename, delimiter='\t')
    df = pd.concat([df, df_one_snr])
df = df.sort_values('SNR')
df
```

```python
plt.plot(df['SNR'], df['Errors'] / df['Total'], label='n0=5, p=128, w=32')
plt.yscale('log')
plt.legend()
plt.grid()
plt.xlabel('SNR')
plt.ylabel('BER')
plt.show()
```
