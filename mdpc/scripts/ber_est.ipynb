```python
# N0 = range(2, 6)
N0 = [5]
P = [128, 256, 512]
for n0 in N0:
    for p in P:
        w = int(32 / 640 * n0 * p)
        df = pd.read_csv('../../results/res_' + str(n0) + str(p) + str(w) + '.txt', delimiter='\t')
        plt.plot(df['SNR'], df['Errors'] / df['Total'], label='n0=' + str(n0) + ' p=' + str(p) + ' w=' + str(w))
        plt.yscale('log')
        plt.legend()
        plt.grid()
        plt.xlabel('SNR')
        plt.ylabel('BER')
        plt.savefig('../../results/fig_' + str(n0) + str(p) + str(w) + '.png')
```
